#### Download, parse (using BeautifulSoup), and print the content from the Trending Developers page from GitHub:

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/developers'
````

````python
response = requests.get(url)
# response.text
````

````python
# your code here
soup = BeautifulSoup(response.content, 'lxml')

names = soup.find_all('h1', attrs = {'class' : 'h3 lh-condensed'})
dev_names = [name.findChild().text.strip('\n').strip() for name in names]
dev_names

links = soup.find_all('p', attrs = {'class' : 'f4 text-normal mb-1'})
dev_links = [link.findChild().text.strip('\n').strip() for link in links]
dev_links

name_links = [name + ' (' +link + ')' for name, link in zip(dev_names, dev_links)]
name_links
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# your code here
response = requests.get(url)
soup = BeautifulSoup(response.content, features = 'lxml')
wikitable = soup.find('table', attrs = {'class' : 'wikitable'})
table_rows = wikitable.find_all('tr')
cols = [th.text.strip('\n') for th in wikitable.find_all('th')]

#QUES - Is it possible to do this without appending?
rows = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip('\n') for tr in td]
    rows.append(row)

    
df = pd.DataFrame(rows[1:], columns = cols)
df.head(10)



````
